```python
from langchain_community.vectorstores import DocArrayInMemorySearch
```

```python
# In memory vector database; ChatGPT will give this context preference when answering the questions
vector_store = DocArrayInMemorySearch.from_texts(
    ["York University has a ML program starting in September 2024", "Population of El Salvador is expected to increase to 8 million by 2026"],
    embedding=OpenAIEmbeddings(openai_api_key = "sk-proj-UMyAS2Pr66RML1aalhQtT3BlbkFJ8DiOWalJnyj28C5ZaMOx")

)
```

```python
retriever = vector_store.as_retriever()
```

```python
# Setup template
template = """
Answer the questions based on the following context:
{context}
Question: {question}
"""
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
llm_model = ChatOpenAI(openai_api_key="sk-proj-UMyAS2Pr66RML1aalhQtT3BlbkFJ8DiOWalJnyj28C5ZaMOx")
```

```python
output_parser = StrOutputParser()
```

```python
setup_and_retrieval = RunnableParallel(
    {"context" : retriever, "question" : RunnablePassthrough()}
    )
```

```python
chain = setup_and_retrieval | prompt | llm_model | output_parser
```

```python
chain.invoke("When is York Universitie's ML course start?")
```

```python
chain.invoke("What is the population of El Salvador?")
```
